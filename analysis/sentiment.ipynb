```python
import pandas as pd
```

```python
kayak_sentiment = pd.read_csv("kayak_roberted.csv")
kayak = pd.read_csv("kayak.csv")
```

```python
kayak.groupby('year')['rating'].mean()
```

```python
kayak_full = kayak.dropna(subset=['review'])
all_reviews = ' '.join(kayak_full['review'].astype(str))
words = all_reviews.lower().split()

word_counts = {}

for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

word_counts_df = pd.DataFrame(list(word_counts.items()), columns=['Word', 'Count'])

sorted_word_counts = word_counts_df.sort_values(by='Count', ascending=False)

sorted_word_counts.to_csv('wordCount.csv', index=False)
```
