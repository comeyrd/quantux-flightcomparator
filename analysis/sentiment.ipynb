```python
import pandas as pd
from tools import words_count
```

```python
def filterthings(reviews_column, possible_words):
    # Convert the reviews column to lowercase for case-insensitive matching
    reviews_lower = reviews_column.str.lower()
    
    # Use list comprehension to filter reviews containing any of the possible words
    filtered_reviews = [review for review in reviews_lower if any(word in review for word in possible_words)]
    
    return filtered_reviews
```
